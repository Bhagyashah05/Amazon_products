```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
product_links = soup.find_all('a', {'class': 'a-link-normal s-no-outline'})
product_urls = []
for link in product_links:
    url = link.get('href')
    if '/dp/' in url:
        product_urls.append('https://www.amazon.in' + url)
print(product_urls)
```

```python
len(product_urls)
```

```python
productname=[]
product_ratings=[]
product_reviews=[]
product_prices=[]
url=[]
des=[]
pasin=[]
pdes=[]
mf=[]
```

```python
for link in product_urls:
  new_webpage=requests.get(link,headers=Headers)
  if new_webpage.status_code==200:
      new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
      description = new_soup.find('meta', {'name': 'description'})['content']
      des.append(description)

      product_description = new_soup.find('div', {'id': 'productDescription'})
      if product_description is not None:
         pdes.append(product_description.text.strip())
      else:
         pdes.append('')

      asin = new_soup.find('th', {'class': 'a-color-secondary a-size-base prodDetSectionEntry'})
      if asin is not None:
         pasin.append(asin.find_next_sibling().text.strip())
      else:
         pasin.append('')

      manufacturer = new_soup.find('a', {'id': 'bylineInfo'})
      if manufacturer is not None:
         mf.append(manufacturer.text.strip())
      else:
         mf.append('')

      url.append(link)
     
      product_name = new_soup.find('span',attrs= {'id': 'productTitle'}).text
      productname.append(product_name)

      rating_elem = new_soup.find('span', {'class': 'a-icon-alt'})
      rating = rating_elem.text.split()[0]
      product_ratings.append(rating)

      reviews_elem = new_soup.find('span', {'id': 'acrCustomerReviewText'})
      reviews = reviews_elem.text.split()[0]
      product_reviews.append(reviews)

      price_span = new_soup.find('span', {'class': 'a-price-whole'})
      if price_span:
        price = float(price_span.text.replace(',', ''))
        product_prices.append(price)
 

```

```python
data={'product_url':url,'product_name':productname,'product_ratings':product_ratings,'product_reviews':product_reviews,'product_price':product_prices,'Description':des,'Asin':pasin,'product_description':pdes,'manufacturer':mf}
df=pd.DataFrame(data)
print(df)
```
